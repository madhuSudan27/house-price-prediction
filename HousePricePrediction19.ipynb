```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
data = remove_outliers_sqft(data)
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
      # store stats of bhk
      bhk_stats = {}
      for bhk, bhk_df in location_df.groupby('bhk'):
        bhk_stats[bhk] = {
            'mean': np.mean(bhk_df.price_per_sqft),
            'std': np.std(bhk_df.price_per_sqft),
            'count': bhk_df.shape[0]
        }

        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:# there should be atleast 5 data point
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
data = bhk_outlier_remover(data)
```

```python
x = data.drop(columns=['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

Apply linear regression


```python
colimn_trans = make_column_transformer((OneHotEncoder(sparse=False),['location']),
                                        remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipeline = make_pipeline(colimn_trans,scaler,lr)
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred_lr = pipeline.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(colimn_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

```python
ridge_regression = Ridge()
```

```python
pipe = make_pipeline(colimn_trans,scaler,ridge_regression)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No regularization", r2_score(y_test,y_pred_lr))
print("Lasso", r2_score(y_test,y_pred_lasso))
print("Ridge", r2_score(y_test,y_pred_ridge))
```

```python
import pickle


# Load the trained model
model = pickle.load(open('RidgeModel.pkl', 'rb'))

# Get the list of locations your model was trained on
known_locations = model.named_steps['columntransformer'].named_transformers_['onehotencoder'].categories_[0]

# Get input from the user
location = input("Enter location: ")
area_type = input("Enter area type: ")
bath = int(input("Enter number of bathrooms: "))
bhk = int(input("Enter number of BHK: "))
total_sqft = float(input("Enter total square feet: "))

# Check if the location is known
if location not in known_locations:
    print("We are not able to predict for this location.")
else:
    # Create a DataFrame from the input
    input_data = pd.DataFrame({
        'location': [location],
        'area_type': [area_type],
        'bath': [bath],
        'bhk': [bhk],
        'total_sqft': [total_sqft]
    })

    # Make the prediction
    predicted_price = model.predict(input_data)[0]

    # Display the predicted price
    print("Predicted price:", predicted_price)
```
